```python
x=df.drop(['CarName','price'],axis=1)
y=df.price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
x_train.select_dtypes(include=object).columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder()
```

```python
oe
```

```python
x_train[['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel',
       'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']]=oe.fit_transform(x_train[['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel',
       'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']])
```

```python
x_test[['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel',
       'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']]=oe.fit_transform(x_test[['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel',
       'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem']])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc
```

```python
sc.fit(x_train)
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
```

```python
lr
```

```python
dtr
```

```python
rfr
```

```python
lr.fit(x_train,y_train)
dtr.fit(x_train,y_train)
rfr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
dtr_pred=dtr.predict(x_test)
rfr_pred=rfr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,lr_pred))
print(r2_score(y_test,dtr_pred))
print(r2_score(y_test,rfr_pred))
```
